```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data.csv')
```

```python
sq=df['sqft_living']
plt.hist(sq,bins=30, color = '#a00309')
plt.title('SqFt Distribution')
plt.xlabel('SqFt')
plt.ylabel('Count')
```

```python
sq=df['sqft_lot']
plt.hist(sq,bins=30, color = '#a00309')
plt.title('Lot SqFt')
plt.xlabel('SqFt')
plt.ylabel('Count')
```

```python
price=df['price']
sq=df['sqft_living']
plt.scatter(x=sq,y=price, color = '#a00309', alpha=.5)
plt.title('Price ~ Living SqFt')
plt.xlabel('SqFt')
plt.ylabel('Price')
```

```python
df=pd.read_excel('../ZestAccuracy.xlsx')
```

```python
df
```

```python
df=df[['Msa', 'Type','Median Error', 'Within 5% of Sale Price']]
```

```python
df['Median Error']=df['Median Error'].str[:-1]
df['Within 5% of Sale Price']=df['Within 5% of Sale Price'].str[:-1]

df['Median Error']=pd.to_numeric(df['Median Error'], errors='coerce')
df['Within 5% of Sale Price']=pd.to_numeric(df['Within 5% of Sale Price'], errors='coerce')
```

```python
df1=df.loc[df.Type=='On-Market']
```

```python

me=df1['Median Error']
plt.hist(me,bins=15, color = '#a00309')
plt.title('On-Market Error Distribution')
plt.xlabel('% Error')
plt.ylabel('Count')
```

```python
df2=df.loc[df.Type=='Off-Market']
```

```python
me=df2['Median Error']
plt.hist(me,bins=15, color = '#a00309')
plt.title('Off-Market Error Distribution')
plt.xlabel('% Error')
plt.ylabel('Count')
```

```python
me=df1['Within 5% of Sale Price']
plt.hist(me,bins=15, color = '#a00309')
plt.title('Within 5% of Sale Price')
plt.xlabel('%')
plt.ylabel('Count')
```
